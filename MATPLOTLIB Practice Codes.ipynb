```python
import matplotlib.pyplot as plt
%matplotlib inline
```

```python
# DATA
import numpy as np
x = np.linspace(0,5,11)
y = x**2
```

```python
x
```

```python
y
```

```python
# Creating A Plot using the FUNCTIONAL METHOD
plt.plot(x,y, 'gp-.')
plt.xlabel('X-axis')
plt.ylabel('Y-axis')
plt.title('Title')
```

```python
# Creating Multiple Plots on the same Canvas
plt.subplot(2,2,1)
plt.plot(x,y, 'rp-.')
plt.subplot(2,2,4)
plt.plot(x,x, 'ko')
```

```python
# Using the Object-Oriented Method
# Creating a Figure and Calling Methods off of it
fig = plt.figure()  # Creates an imaginary Figure

# Adding axes to the Figure
axes = fig.add_axes([0.1,0.1,0.8,0.8])

# Plotting The axes
axes.plot(x,y, 'y-..')

# Labelling the axes
axes.set_xlabel('X-axis')
axes.set_ylabel('Y-axis')
axes.set_title('Figure PLot')
```

```python
# Multiple axis
fig = plt.figure()
axes1 = fig.add_axes([0.1,0.1,0.8,0.8])
axes2 = fig.add_axes([0.4,0.2,0.4,0.3])
axes1.plot(y,x, 'kd-.')
axes2.plot(x,y, 'r.-.')
axes1.set_title('Larger PLot')
axes2.set_title('Smaller Plot')
```

# Creating Subplots using the Object-Oriented Method


```python
# A one by one Subplot
fig, axes = plt.subplots()
axes.plot(x,y, 'kd')
```

```python
fig, axes = plt.subplots(4,4)
plt.tight_layout() # fixes the overlaps
```

```python
fig,axes = plt.subplots(1,2)

# Plotting on the subplots through indexing
axes[0].plot(x,y, 'm')
axes[1].plot(y,x, 'k')

# Setting Titles
axes[0].set_title('1st Plot')
axes[1].set_title('2nd Plot')

fig.tight_layout()
```

```python
fig,axes = plt.subplots(1,2)

# plotting through iteration
for ax in axes:
    ax.plot(x,y, 'orange')
    ax.set_xlabel('x')
    ax.set_ylabel('y')
    ax.set_title('Plot')  
    
fig.tight_layout() # Used Interchangeably with plt.tight_layout()
```

```python
# Figure Size and DPI
fig = plt.figure(figsize = (3,2), dpi = 200)
ax = fig.add_axes([0,0,1,1])
ax.plot(x,y, 'k')
ax.set_title('Title')
ax.set_xlabel('X-Label')
ax.set_ylabel('Y-Label')
```

```python
# Saving Figure to a File
fig.savefig('figure_sweetie.jpg', dpi = 100) # Automaatically Saves the Figure 
```

```python
# LEGENDS - Used to Clarify What Plot is What Plot
fig = plt.figure()
ax = fig.add_axes([0,0,1,1])
ax.plot(x,x*3, label = 'X-Cubed')
ax.plot(x,x**2, label = 'X-Squared')
ax.legend(loc = 0)
```

```python
fig = plt.figure()
ax = fig.add_axes([0,0,1,1])
ax.plot(x,x*3, label = 'X-Cubed')
ax.plot(x,x**2, label = 'X-Squared')
ax.legend(loc = (0.1,0.8)) # Tuple Specifying where I want the Location of the LEGEND to be
```

```python
# Setting Colors,Line Widths and Line Styles
# Setting Colors
fig = plt.figure()
ax = fig.add_axes([0,0,1,1])
ax.plot(x,y, color = 'purple')
```

```python
# Setting Colors Using Hexadecimal String
fig = plt.figure()
ax = fig.add_axes([0,0,1,1])
ax.plot(x,y, color = '#F89C11')
```

```python
# Line Width,short form is lw
fig = plt.figure()
ax = fig.add_axes([0,0,1,1])
ax.plot(x,y, color = 'purple', lw = 6)
```

```python
# Alpha  Argument which allows one to basically control how transparent the line is
fig = plt.figure()
rand_ax= fig.add_axes([0.4,0.6,0.5,0.8])
rand_ax.plot(x,y, 'k', lw = 6, alpha = 0.2)
```

```python
# Linestyles or ls for short
fig = plt.figure()
rand_ax= fig.add_axes([0.4,0.6,0.5,0.8])
rand_ax.plot(x,y, 'k', ls ='dotted')
```

```python
# Markers - used when one has a few datapoints
fig = plt.figure()
rand_ax= fig.add_axes([0.4,0.6,0.5,0.8])
rand_ax.plot(x,y, 'k', ds = 'steps', marker = 'o', markersize = 14, markerfacecolor = 'red', markeredgewidth = 4, markeredgecolor = 'k')
```

```python
# Control Over Axes Appearance
fig = plt.figure(figsize = (3,2))
ax = fig.add_axes([0,0,1,1])
ax.plot(x,y, 'k')
ax.set_title('Title')
ax.set_xlabel('X-Label')
ax.set_ylabel('Y-Label')

ax.set_xlim([0,2])
ax.set_ylim([0,3])
```

```python
# Special Plot Types
# Scatter Plot
plt.scatter(x,y)
```
